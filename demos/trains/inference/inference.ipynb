# Inference - Query the base and new model

```python
import cv2
from matplotlib import pyplot as plt
from PIL import Image
## note that to keep the notebook readable, a lot of code was moved into the file remote_infer.py.
## if you are curious, open it to see the various functions required around this prediction
from utils import *
```

## Variable declaration

```python
# New model endpoint (model that detects lego traffic signs)
RestUrl = 'http://modelmesh-serving:8008' # CHANGEME with your endpoint
# Base model endpoint
base_model_endpoint = 'http://modelmesh-serving.base-model:8008'
# Example image
image_path = 'images/inference.jpg'
```

## Base model detection

```python
# The base model endpoint
base_model_infer_url = f'{base_model_endpoint}/v2/models/base/infer'
# Launch the inference by calling the remote model server
img = process_image(image_path, base_model_infer_url)

# Display the result
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
fig = plt.gcf()
#fig.set_size_inches(6, 3)
plt.axis('off')
plt.imshow(img);
```

## New model detection

```python
# The new model endpoint
new_model_infer_url = f'{RestUrl}/v2/models/new/infer'
# Launch the inference by calling the remote model server
img = process_image(image_path, new_model_infer_url)

# Display the result
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
fig = plt.gcf()
#fig.set_size_inches(6, 3)
plt.axis('off')
plt.imshow(img);
```
